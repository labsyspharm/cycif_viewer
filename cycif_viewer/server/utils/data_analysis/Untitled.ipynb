```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(4, 1, figsize=(10,40))
i = 0
for metric in ['cosine']:
    for n_neighbors in [100, 200]:
        for min_dist in [.2, .8]:
            save_string = 'full_weighted_umap_' + metric + '_' + str(n_neighbors) + 'neighbors_' + str(min_dist) + 'dist'
            save_string = save_string.replace('.', '')
            A = np.load(save_string+'.npy')
            axs[i].scatter(A[:, 0], A[:, 1], s=0.1, alpha=0.1)
            axs[i].set_title(save_string)
            i+=1

            

```

```python
from sklearn.cluster import KMeans

fig, axs = plt.subplots(1, 1, figsize=(10,10))
i = 0
for metric in ['cosine']:
    for n_neighbors in [ 200]:
        for min_dist in [.2]:
            save_string = 'full_weighted_umap_' + metric + '_' + str(n_neighbors) + 'neighbors_' + str(min_dist) + 'dist'
            save_string = save_string.replace('.', '')
            A = np.load(save_string+'.npy')
            y_kmeans = KMeans(random_state=0).fit_predict(A)
            axs.scatter(A[:, 0], A[:, 1], c=y_kmeans, s=0.1, alpha=0.1, cmap='viridis')
            axs.set_title(save_string+"kmeans")
            i+=1
```
